```python
%%bash
cd Headers/
cat > seqMeasurer
#!/bin/bash
>lengths.txt
for file in *.fa; do
     grep ">" "$file" >> lengths.txt
     echo -n "File: $file , Total nucleotides:" >> lengths.txt
     grep -v ">" "$file" | wc -m >> lengths.txt

done
```

```python
%%bash
cd Headers/
chmod u+x seqMeasurer
```

```python
%%bash
cd Headers/
./seqMeasurer
```
